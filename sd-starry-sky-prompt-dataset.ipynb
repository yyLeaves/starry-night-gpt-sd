```python
prompts = pd.read_csv('/kaggle/input/openprompts/prompts.csv').drop('raw_data', axis = 1)
prompts = prompts.loc[prompts.prompt.str.split(' ').str.len() > 6]
prompts = prompts.reset_index(drop=True)
prompts
```

```python
list_prompts = [p for p in prompts['prompt'] if 'star' in p and 'sky' in p and 'night' in p]
```

```python
for i in range(5):
    print(f"{i}")
    print(random.choice(list_prompts))
    print()
```

```python
with open('starry_night_prompts.txt', 'w') as f:
    f.write('\n\n'.join(list_prompts))
```
